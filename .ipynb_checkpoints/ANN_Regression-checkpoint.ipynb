```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset.isnull().any()
```

```python
dataset["Administration"].fillna(dataset["Administration"].mode()[0], inplace=True)
dataset["Marketing Spend"].fillna(dataset["Marketing Spend"].mode()[0], inplace=True)
```

```python
dataset.head(12)
```

```python
x = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4:5].values
```

```python
x
```

```python
y
```

```python
dataset.head()
```

```python
dataset["State"].unique()
```

```python
# 0-California, 1-Florida, 2-New York
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("oh", OneHotEncoder(), [3])],remainder="passthrough")
x = ct.fit_transform(x)
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
regression = Sequential()
```

```python
regression.add(Dense(units = 6, kernel_initializer = "random_uniform",activation = "relu")) #x_train, 1st hidden layer
```

```python
regression.add(Dropout(0.2))
```

```python
regression.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu")) #x_train, 2nd hidden layer
```

```python
regression.add(Dropout(0.2))
```

```python
regression.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu")) #x_train, 3rd hidden layer
```

```python
regression.add(Dropout(0.2))
```

```python
regression.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu")) #x_train, 4th hidden layer
```

```python
regression.add(Dense(units = 1, kernel_initializer = "random_uniform",activation = "relu")) #x_train
```

```python
regression.compile(optimizer = "adam", loss = "mse", metrics = ['mse'])
```

```python
history = regression.fit(x_train,y_train, batch_size=50, epochs = 800)
```

```python
ypred = regression.predict(x_test)
```

```python
ypred
```

```python
ypred>0.5
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,ypred)
```
