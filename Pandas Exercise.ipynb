**MERGING**

```python
df1 = pd.DataFrame({'Name':['Meredith','Cristina','Alex','Jackson','April'], 'Age':[32,31,29,27,26], 'Gender':['F','F','M','M','F']})
df2 = pd.DataFrame({'Name':['Derek','Owen','Jo','Lexie','Mark'], 'Age':[32,31,29,27,26], 'Gender':['M','M','F','F','M']})
mer = pd.merge(df1,df2,on='Age')
mer
```

**CONCATENATION**

```python
concat = pd.concat([df1,df2])
concat
```

```python
concat = pd.concat([df1,df2],axis=1)
concat
```

**JOINING**

```python
df1 = pd.DataFrame({'Name':['Meredith','Cristina','Alex','Jackson','April'], 'Age':[32,31,29,27,26], 'Gender':['F','F','M','M','F']})
df3 = pd.DataFrame({'City':['Mumbai','Delhi','Hyderabad','Bangalore','Chennai'], 'State':['MAH','DEL','TEL','KN','TN']})
join = df1.join(df3)
join
```
