```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(3,5,5)
y = x**3
```

```python
x
```

```python
y
```

```python
plt.xlim(3,5)
plt.ylim(3**3,5**3)
plt.plot(x,y,color="red")
plt.scatter(x,y,color="blue")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.show()
```

```python
plt.plot(x,y,color="red",label="x vs y")
plt.plot(y,x,color="blue",label="y vs x")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.legend()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,color="red")
plt.subplot(2,2,2)
plt.scatter(x,y,color="blue")
```

```python
plt.bar(x,y,color="red")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.show()
```

```python
sports = ["Football","Basketball","Cricket","Badminton","Hockey"]
people = [40,30,70,25,20]
```

```python
plt.pie(people, labels=sports, autopct='%0.1f%%', colors=["red","pink","yellow","lightblue","silver"])
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(x,y,color="red")
ax2.scatter(x,y,color="blue")
ax3.bar(x,y,color="lime")
ax4.pie(people, labels=sports, autopct='%0.1f%%', colors=["red","pink","yellow","lightblue","silver"])
plt.show()
```
