```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("oh", OneHotEncoder(), [0])],remainder="passthrough")
y = ct.fit_transform(y).toarray()
```

```python
y
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
categorical = Sequential()
```

```python
categorical.add(Dense(units = 6, kernel_initializer = "random_uniform",activation = "relu")) #x_train
```

```python
categorical.add(Dropout(0.2))
```

```python
categorical.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu"))
```

```python
categorical.add(Dropout(0.2))
```

```python
categorical.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu"))
```

```python
categorical.add(Dropout(0.2))
```

```python
categorical.add(Dense(units = 12, kernel_initializer = "random_uniform",activation = "relu"))
```

```python
categorical.add(Dense(units = 7, kernel_initializer = "random_uniform",activation = "softmax")) #y_train
```

```python
categorical.compile(optimizer = "rmsprop", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
history = categorical.fit(x_train,y_train, batch_size=64, epochs = 1000, validation_data = (x_test,y_test))
```

training accuracy is very high and testing is low = Over fitting

training accuracy is low and testing is high = Under fitting

training accuracy and training is approximately = good
### Fine Tuning Criteria
1. Change the random_state to some other value
2. Add hidden layers
3. Adding drop out layers
4. Adding your custom learning rate
5. Changing the batch size as well as optimizer
6. Adding some more data

Different optimizers, other than gradient descent, rmsprop and difference between them

<0.5 - Worst Model

>0.5 to <0.6 - Below Avg Model

>0.6 to <0.7 - Avg Model

>0.7 to <0.8 - Above Avg Model

>8 to <8.7 - Good Model

>8.5 to <9.5 - Best Model
