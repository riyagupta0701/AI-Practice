```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(11,22,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.xlim(11,22)
plt.ylim(11**2,22**2)
plt.plot(x,y,color="red")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.show()
```

```python
plt.xlim(11,22)
plt.ylim(11**2,22**2)
plt.scatter(x,y,color="red")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.show()
```

```python
# plt.xlim(11,22)
# plt.ylim(11**2,22**2)
plt.plot(x,y,color="red",label="x vs y")
plt.plot(y,x,color="green",label="y vs x")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.legend(loc="center")
plt.show()
```

```python
plt.subplot(2,2,1)       #plt.subplot(rows,columns,division number)
plt.scatter(x,y,color="red")
plt.subplot(2,2,4)
plt.plot(x,y,color="purple")
```

```python
plt.bar(x,y,color="red")
plt.xlabel("This is x label")
plt.ylabel("This is y label")
plt.title("This is title")
plt.show()
```

```python
courses = ["AI/ML","IOT","Blockchain","Python"]
students = [200,175,180,120]
```

```python
plt.pie(students, labels=courses, autopct='%0.1f%%', colors=["crimson","pink","yellow","orange"])
plt.show()
```

```python
fig = plt.figure(figsize = (20,20))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.plot(x,y,color="red")
ax2.scatter(x,y,color="green")
ax3.bar(x,y,color="blue")
ax4.pie(students, labels=courses, autopct='%0.1f%%', colors=["crimson","pink","yellow","orange"])
plt.show()
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
sns.lineplot(x="size", y="tip", data=tips)
```

```python
sns.lineplot(x="size", y="tip", ci=50, data=tips)    #ci: confidence interval
```

```python
sns.distplot(tips["tip"])
```

```python
sns.barplot(x="day", y="tip", data=tips)
```

```python
sns.boxplot(x="size", y="tip", data=tips)
```

```python
tips.describe()
```

```python
sns.catplot(x="smoker", y="tip", hue="sex", data=tips)
```

```python
sns.catplot(x="day", y="tip", hue="sex", kind="bar",color="red", data=tips)
```

```python
sns.pairplot(tips)     #only for numerical data
```
